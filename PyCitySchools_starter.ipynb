```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

```python
#Printing out merged data head
school_data_complete.head()
```

```python
#Total number of schools
schoolnames = school_data_complete["school_name"].unique()
schoolnamecount=len(schoolnames)
schoolnamecount
```

```python
#Total number of students
studentcount = len(school_data_complete["Student ID"])
studentcount
```

```python
#Total budget
totalbudget = school_data["budget"].sum()
totalbudget
```

```python
#Average math score
avg_mathscore = school_data_complete["math_score"].mean()
avg_mathscore
```

```python
#Average reading score
avg_readingscore = school_data_complete["reading_score"].mean()
avg_readingscore
```

```python
#Percent students with passing math score (>=70)
passingmath = school_data_complete.loc[school_data_complete["math_score"] >= 70]
percentpassingmath = (len(passingmath)/studentcount)*100
percentpassingmath
```

```python
#Percent students with passing reading score (>=70)
#Percent students with passing math score (>=70)
passingreading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
percentpassingreading = (len(passingreading)/studentcount)*100
percentpassingreading
```

```python
#Percent students with passing math AND reading score (% Overall Passing)
passingboth = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
percentpassingboth = (len(passingboth)/studentcount)*100
percentpassingboth
```

```python
#Dataframe to hold above results
districtsummary_df = pd.DataFrame(
    {"Total Schools": [schoolnamecount],
    "Total Students": [studentcount],
    "Total Budget": [totalbudget],
    "Average Math Score": [avg_mathscore],
    "Average Reading Score": [avg_readingscore],
    "% Passing Math": [percentpassingmath],
    "% Passing Reading": [percentpassingreading],
    "% Overall Passing": [percentpassingboth]
})

districtsummary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School type
schooltype = school_data_complete.groupby("school_name")['type'].unique()
schooltype
```

```python
#Total Students
totalstudents = school_data_complete.groupby('school_name')['Student ID'].nunique()
totalstudents
```

```python
#Total School Budget
budget = school_data_complete.groupby('school_name')['budget'].sum()
totalschoolbudget = budget/totalstudents
totalschoolbudget
```

```python
#Per Student Budget
studentbudget = totalschoolbudget/totalstudents
studentbudget
```

```python
#Average Math Score
avgmathscore = school_data_complete.groupby('school_name')['math_score'].mean()
avgmathscore
```

```python
#Average Reading Score
avgreadingscore = school_data_complete.groupby('school_name')['reading_score'].mean()
avgreadingscore
```

```python
#Calculating percentage math pass rate by school
schoolpassmath = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()/totalstudents*100
schoolpassmath
```

```python
#% Passing Reading
schoolpassread = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()/totalstudents*100
schoolpassread
```

```python
#% Overall Passing
schoolpassboth = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("school_name")["student_name"].count()/totalstudents*100
schoolpassboth
```

```python
#Merging series
merge_df = pd.merge(schooltype, totalstudents, on="school_name")
merge_df
```

```python
merge2_df = pd.merge(merge_df, avgmathscore, on="school_name")
merge2_df
```

```python
merge3_df = pd.merge(merge2_df, avgreadingscore, on="school_name")
merge3_df
```

```python
#For some reason, (possibly because of division by a groupby in totalstudents variable), these dataframes will not merge with the others directly. So, created a new dataframe to hold these.
dividedbytotalstudents_df=pd.DataFrame ({"Total School Budget": totalschoolbudget,
                          "Per Student Budget": studentbudget,
                          "% Passing Math": schoolpassmath,
                          "% Passing Reading": schoolpassread,
                          "% Overall Passing": schoolpassboth
                            })
```

```python
#Merging all series into single dataframe
all_df = pd.merge(merge3_df, dividedbytotalstudents_df, on="school_name")
all_df
```

```python
#School Summary renamed
schoolsummary_df = all_df.rename(columns={"type":"School Type", "Student ID":"Total Students","math_score":"Average Math Score","reading_score":"Average Reading Score"})
schoolsummary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Finding the max and min values of the dataframe to determine appropriate bins
print(all_df["Per Student Budget"].max())
print(all_df["Per Student Budget"].min())
```

```python
#Creating bins based on max/min student budget values
bins = [575,597.25,616.5,635.75,655]
#bin names
ranges = ["$575-$597.25", "$597.25-$616.5","$616.5-$635.75", "$635.75-$655"]
```

```python
#Slicing the data and placing into bins
pd.cut(schoolsummary_df["Per Student Budget"], bins, labels=ranges)
```

```python
#Creating a column for spending ranges
schoolsummary_df["Spending Ranges (Per Student)"] = pd.cut(schoolsummary_df["Per Student Budget"], bins=bins, labels=ranges)
schoolsummary_df.reset_index(drop=True,inplace=True)
```

```python
#Creating series for each parameter. Need to take the average for each range of school spending!
mathscorespend = schoolsummary_df.groupby("Spending Ranges (Per Student)")["Average Math Score"].mean()
readscorespend = schoolsummary_df.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()
percentmathspend = schoolsummary_df.groupby("Spending Ranges (Per Student)")["% Passing Math"].mean()
percentreadspend = schoolsummary_df.groupby("Spending Ranges (Per Student)")["% Passing Reading"].mean()
percentoverallspend = schoolsummary_df.groupby("Spending Ranges (Per Student)")["% Overall Passing"].mean()
```

```python
#Create dataframe to hold data
scoresbyspending_df = pd.DataFrame(
    {"Average Math Score": mathscorespend,
    "Average Reading Score": readscorespend,
    "% Passing Math": percentmathspend,
    "% Passing Reading": percentreadspend,
    "% Overall Passing": percentoverallspend})

scoresbyspending_df
```
